This notebook generates synthetic datasets for a subscription-based business and performs basic cleaning & preprocessing.


```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta

fake = Faker()
```

## 1. Generate Customers Dataset

```python
num_customers = 20000
start_date = datetime(2023,1,1)
end_date = datetime(2024,12,31)

customers=[]
channels=["Ads","Referral","Organic","Email","Social"]
genders=["Male","Female","Other"]
countries=["India","USA","UK","Canada","UAE"]

for i in range(1, num_customers+1):
    customers.append([
        i,
        fake.name(),
        random.choice(genders),
        random.randint(18,60),
        random.choice(countries),
        fake.city(),
        fake.date_between(start_date, end_date),
        random.choice(channels)
    ])

customers_df = pd.DataFrame(customers, columns=[
    "customer_id","customer_name","gender","age","country",
    "city","registration_date","acquisition_channel"
])
```

## 2. Save Dataset

```python
customers_df.to_csv("customers_demo.csv", index=False)
customers_df.head()
```
